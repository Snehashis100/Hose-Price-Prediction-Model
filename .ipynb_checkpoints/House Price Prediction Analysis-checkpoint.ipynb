```python
def IQR(df,feature):
    outliers=[]
    q1,q3=np.percentile(df[feature],[25,75])
    lower=q1 - 3*(q3-q1)
    upper=q3 + 3*(q3-q1)
    for i in df[feature]:
        if i<lower or i>upper:
            outliers.append(i)
    return outliers
```

```python
print(f"Outliers of Price: {len(IQR(data1,'price'))}") # Checking for extreme outliers
```

```python
def Outlier_Record(df,feature,outlier):
    lst=[]
    for i in range(len(df[feature])):
        if df.loc[i,feature] in outlier:
            lst.append(df.loc[i,:])
    record=pd.DataFrame(lst,columns=df.columns)
    return record
```

```python
# Printing the correlation of the independent features with the depending feature in descending order
correlation=[]
for feature in data6.columns:
    if feature!='price':
        corr_val=data6[['price',feature]].corr().values[0,1]
        correlation.append(corr_val)

a=list(data6.columns)
a.remove('price')
x=pd.DataFrame(np.array(correlation).reshape(-1,1),columns=['Correlation'])
x.index=a
x.nlargest(len(data6.columns),columns='Correlation')
```

```python
x=data6.drop(['price'],axis=1)
y=data6['price']
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_train2=mms.fit_transform(X_train)
X_test2=mms.transform(X_test)
X_train=pd.DataFrame(X_train2,columns=X_train.columns)
X_test=pd.DataFrame(X_test2,columns=X_test.columns)
```

```python
from sklearn.metrics import r2_score
```

```python
def Model_Performance(x_train,x_test,y_train,y_test,model,r2_score):
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    r2=r2_score(y_test,pred)
    adj_r2 = 1 - (1-r2)*(len(x_test) - 1) / (len(x_test) - (x_test.shape[1] - 1) - 1)
    return r2,adj_r2
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
from sklearn.svm import SVR
import xgboost as xgb
```

```python
# Linear Regression
linear=LinearRegression()
linear_r2,linear_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,linear,r2_score)
print(f"r2 score of Linear Regression :{linear_r2}")
print(f"Adjusted r2 score of Linear Regression :{linear_adj_r2}")
```

```python
# Ridge Regression
ridge=Ridge()
ridge_r2,ridge_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,ridge,r2_score)
print(f"r2 score of Ridge Regression :{ridge_r2}")
print(f"Adjusted r2 score of Ridge Regression :{ridge_adj_r2}")
```

```python
# Lasso Regression
lasso=Lasso()
lasso_r2,lasso_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,lasso,r2_score)
print(f"r2 score of Lasso Regression :{lasso_r2}")
print(f"Adjusted r2 score of Lasso Regression :{lasso_adj_r2}")
```

```python
# Random Forest Regression
rf=RandomForestRegressor()
rf_r2,rf_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,rf,r2_score)
print(f"r2 score of Random Forest Regression :{rf_r2}")
print(f"Adjusted r2 score of Random Forest Regression :{rf_adj_r2}")
```

```python
# Desision Tree Regression
dt=DecisionTreeRegressor()
dt_r2,dt_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,dt,r2_score)
print(f"r2 score of Decision Tree Regression :{dt_r2}")
print(f"Adjusted r2 score of Decision Tree Regression :{dt_adj_r2}")
```

```python
# Gradient Boost Regression
gb=GradientBoostingRegressor()
gb_r2,gb_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,gb,r2_score)
print(f"r2 score of Gradient Boost Regression :{gb_r2}")
print(f"Adjusted r2 score of Gradient Boost Regression :{gb_adj_r2}")
```

```python
# Ada Boost Regression
ada=AdaBoostRegressor()
ada_r2,ada_adj_r2=Model_Performance(X_train,X_test,y_train,y_test,ada,r2_score)
print(f"r2 score of Gradient Boost Regression :{ada_r2}")
print(f"Adjusted r2 score of Gradient Boost Regression :{ada_adj_r2}")
```

```python
# Trying ANN
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
```

```python
# Hyperparameter tuning for getting the best ANN with all the appropriate values for no of hidden layers ,no of neurons and the 
# learning rate
def hyper_model(hp):
    model=keras.Sequential() # Building the ANN
    for i in range(hp.Int('num_layers',2,20)):  # ANN must have hidden layers in between 2 to 20
        model.add(layers.Dense(units=hp.Int('units_'+str(i),min_value=32,max_value=300,step=10),# In each hidden layers the no of neurons must be between 32 to 512
                              activation='relu')) # activation function used
        model.add(layers.Dense(1,activation='linear'))# Since it is a regressiom problem the output neuron should be 1 and the activation function should be linear
        model.compile(keras.optimizers.Adam(      # Using Adam as optimizer and hypertune its parameter
                      hp.Choice('learning_rate',[1e-1,1e-2, 1e-3, 1e-4])), # LR for Adam
                      loss='mean_absolute_error',
                      metrics=['mean_absolute_error'])
    return model
```

```python
tuner = RandomSearch(
    hyper_model,
    objective='val_mean_absolute_error', # metric used for evaluating the ANN
    max_trials=5,
    executions_per_trial=3,
    directory='project',
    project_name='log')# there will be 5 trials and inside each trial there will be 3 execution so in total 15 trials
```

```python
tuner.search_space_summary()
```

```python
tuner.search(X_train, y_train,epochs=5,validation_data=(X_train, y_train))
```

```python
tuner.results_summary()
```
